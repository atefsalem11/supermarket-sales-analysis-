## Data Cleaning

```python
# Check missing values
df.isnull().sum()
```

```python
### Data type conversion
df['Date'] = pd.to_datetime(df['Date'])
# Convert Time column to datetime objects, handling the inconsistent format
df['Time'] = pd.to_datetime(df['Time'].str.replace(' - ', ':'), format='%I:%M %p', errors='coerce').dt.time
df['Unit price'] = pd.to_numeric(df['Unit price'], errors='coerce')
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')
```

## Feature Engineering

```python
# Create new features
df['Month'] = df['Date'].dt.month
df['Day_of_week'] = df['Date'].dt.day_name()
df['Revenue_per_item'] = df['Unit price'] * df['Quantity']
df['Profit_margin'] = df['Revenue_per_item'] * 0.2
```

## Sales Analysis

```python
# Branch analysis
branch_sales = df.groupby('Branch')['Total'].sum()
print("Sales by Branch:")
print(branch_sales)
```

```python
# Product line analysis
product_sales = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print("Sales by Product Line:")
print(product_sales)
```

## Visualizations

```python
# Monthly Sales Trend
plt.figure()
monthly_sales = df.groupby('Month')['Total'].sum()
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linewidth=2)
plt.title('Monthly Sales Trend', pad=20, size=14)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Product Line Sales Distribution
plt.figure()
sns.barplot(data=df, x='Product line', y='Total')
plt.xticks(rotation=45)
plt.title('Sales Distribution by Product Line', pad=20, size=14)
plt.tight_layout()
plt.show()
```

```python
# Customer Ratings Distribution
plt.figure()
sns.histplot(data=df, x='Rating', bins=20)
plt.title('Distribution of Customer Ratings', pad=20, size=14)
plt.show()
```

## Key Findings

```python
def get_key_metrics(df):
    metrics = {
        'Total Revenue': f"${df['Total'].sum():,.2f}",
        'Average Transaction': f"${df['Total'].mean():,.2f}",
        'Average Rating': f"{df['Rating'].mean():,.2f}",
        'Most Popular Product': df.groupby('Product line')['Quantity'].sum().idxmax(),
        'Best Branch': df.groupby('Branch')['Total'].sum().idxmax(),
        'Popular Payment': df['Payment'].mode()[0]
    }
    return pd.Series(metrics)

print("Key Business Metrics:")
print("-" * 40)
print(get_key_metrics(df))
```
